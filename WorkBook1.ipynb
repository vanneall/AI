## Пример 4.2.2
Для функции sin(𝑥𝑒^cos(𝑥)) на интервале [0;10] с шагом 0,05 построить
график функции, закрасить площадь под ним и вычислить значение
этой площади.
Для вычисления площади воспользуйте функции trapz(), вычисляющей
интеграл по правилу трапеции. Для ее корректной работы необходимо
подключить следующие библиотеки:
from scipy.integrate import simps
from numpy import trapz

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz

x = np.arange(0.0, 10, 0.1)
y = np.abs(np.sin(x * np.exp(np.cos(x))))
plt.grid()
plt.plot(x, y, c='r')
plt.fill_between(x, y)

area = trapz(y)
print(area)
```

## Задание 4.3.2
Дана функция. Создать массив из 10 значений
функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой
половины массива и построить графики для основного массива –
линейный и для среза – точечный


```python
import math
import numpy as np
import matplotlib.pyplot as plt

list1 = []
for x in range(1, 11):
    a = math.sqrt(1 + math.exp(math.sqrt(x)) + math.cos(x ^ 2))
    b = math.fabs(1 - math.pow(math.sin(x), 3))
    c = a / b
    d = math.log(math.fabs(2 * x))
    res = round(c + d, 2)
    list1.append(res)
print(list1)
list2 = list1[:5]

x1 = range(1, 6, 1)
y1 = list2
plt.scatter(x1, y1)
plt.show()

x2 = range(1, 11, 1)
y2 = list1
plt.plot(x2, y2)
```

## Задание 4.3.3
Дана функция. Построить график на
интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь под ним. Для вычисления площади воспользуйте функции trapz(),
вычисляющей интеграл по правилу трапеции.
Для ее корректной работы необходимо подключить следующие
библиотеки:
from scipy.integrate import simps
from numpy import trapz

```python
import math
import numpy as np
import matplotlib.pyplot

list1 = []
for x in range(1, 11):
    a = math.exp(math.cos(x) + math.log(1 + x))
    res = round(math.fabs(math.cos(x * a)), 2)
    list1.append(res)
print(list1)

x2 = range(1, 11, 1)
y2 = list1
matplotlib.pyplot.plot(x2, y2)
matplotlib.pyplot.fill_between(x2, y2)
print(np.trapz(list1, range(1, 11)))
```

## Задание 4.3.5
Создайте простейший калькулятор, включающий основные действия
для двух переменных ′ + ′, ′ − ′, ′ ∙ ′, ′ ÷ ′, а также вычисление
следующих функций: 𝑒^(𝑥+𝑦), sin(𝑥 + 𝑦) , cos(𝑥 + 𝑦) , 𝑥^𝑦.

```python
import math

print("Конец работы - 0")
while True:
    sign = input("Знак (+, -, *, /, sin, cos, e, pow): ")
    if sign == '0':
        break
    x = float(input("x="))
    y = float(input("y="))
    if sign == '+':
        print(round((x + y), 2))
    elif sign == '-':
        print(round((x - y), 2))
    elif sign == '*':
        print(round((x * y), 2))
    elif sign == '/':
        if y != 0:
            print("%.2f" % (x / y))
        else:
            print("нельзя делить на 0")
    elif sign == 'sin':
        print(math.sin(x + y))
    elif sign == 'cos':
        print(math.cos(x + y))
    elif sign == 'e':
        print(math.exp(x + y))
    elif sign == 'pow':
        print(math.pow(x, y))
    else:
        print("Неверный знак операции!")
        
```
