## –ó–∞–¥–∞–Ω–∏–µ 4.3.2
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è. –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π


```python
import math
import numpy as np
import matplotlib.pyplot as plt

list1 = []
for x in range(1, 11):
    a = math.sqrt(1 + math.exp(math.sqrt(x)) + math.cos(x ^ 2))
    b = math.fabs(1 - math.pow(math.sin(x), 3))
    c = a / b
    d = math.log(math.fabs(2 * x))
    res = round(c + d, 2)
    list1.append(res)
print(list1)
list2 = list1[:5]

x1 = range(1, 6, 1)
y1 = list2
plt.scatter(x1, y1)
plt.show()

x2 = range(1, 11, 1)
y2 = list1
plt.plot(x2, y2)
```

–≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.

–±–∏–±–ª–∏–æ—Ç–µ–∫–∏:
from scipy.integrate import simps
from numpy import trapz

```python
import math
import numpy as np
import matplotlib.pyplot

list1 = []
for x in range(1, 11):
    a = math.exp(math.cos(x) + math.log(1 + x))
    res = round(math.fabs(math.cos(x * a)), 2)
    list1.append(res)
print(list1)

x2 = range(1, 11, 1)
y2 = list1
matplotlib.pyplot.plot(x2, y2)
matplotlib.pyplot.fill_between(x2, y2)
print(np.trapz(list1, range(1, 11)))
```

```python
import math

print("–ö–æ–Ω–µ—Ü —Ä–∞–±–æ—Ç—ã - 0")
while True:
    sign = input("–ó–Ω–∞–∫ (+, -, *, /, sin, cos, e, pow): ")
    if sign == '0':
        break
    x = float(input("x="))
    y = float(input("y="))
    if sign == '+':
        print(round((x + y), 2))
    elif sign == '-':
        print(round((x - y), 2))
    elif sign == '*':
        print(round((x * y), 2))
    elif sign == '/':
        if y != 0:
            print("%.2f" % (x / y))
        else:
            print("–Ω–µ–ª—å–∑—è –¥–µ–ª–∏—Ç—å –Ω–∞ 0")
    elif sign == 'sin':
        print(math.sin(x + y))
    elif sign == 'cos':
        print(math.cos(x + y))
    elif sign == 'e':
        print(math.exp(x + y))
    elif sign == 'pow':
        print(math.pow(x, y))
    else:
        print("–ù–µ–≤–µ—Ä–Ω—ã–π –∑–Ω–∞–∫ –æ–ø–µ—Ä–∞—Ü–∏–∏!")
        
```
